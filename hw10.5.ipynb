```python
from scipy.stats import t, norm
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

rv1 = t(3)

def graphs(a):
        
        rv2 = norm(0,1)
        df1 = pd.DataFrame({})
        for i in range (1,10000):
            df1[i] = rv1.rvs(size = a)*np.sqrt(a/3) #нормируем домножением на корень

        df1.loc["Mean"] = df1.mean()
        print(df1)
        
        df1.loc["PDF"] = rv2.pdf(df1.loc["Mean"]) #ищем плотность

        plt.tight_layout()
        
        sns.histplot(x = df1.loc["Mean"], color = "Pink", stat = "density", binrange = (-10,10))
        sns.lineplot( x = df1.loc["Mean"], y = df1.loc["PDF"], color = "Orange")

graphs(2)
```

```python
graphs(5)
```
